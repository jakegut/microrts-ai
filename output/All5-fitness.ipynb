```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("fitness-finalfinal.txt")
# print(-df[df['evolution'] != 1])

df[df['evolution'] != 1] = -df[df['evolution'] != 1]

df['iteration'] = df['iteration'].abs()
df['evolution'] = df['evolution'].abs()

print(df)
```

```python
draw_df = df.reset_index().melt(id_vars=['index', 'iteration', 'evolution'], var_name='col')
print(draw_df)
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(14,7))
g = sns.lineplot(x="iteration", y="value", data=draw_df)
plt.xlabel("Iteration", fontsize=24)
plt.ylabel("Fitness", fontsize=24)
# plt.title("Mean Fitness per iteration for 5 evolutions", fontsize=20)
for label in (g.get_xticklabels() + g.get_yticklabels()):
	label.set_fontsize(22)
plt.show(g)
```
