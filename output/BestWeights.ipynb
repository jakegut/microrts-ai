```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("best-weights.txt")
print(df.head())
```

```python
draw_df = df.reset_index().melt(id_vars=['index', 'evolution'], var_name='function').pivot("function", "evolution", "value")

sns.set(font_scale=1.5)
plt.figure(figsize=(7,7))
g = sns.heatmap(draw_df, annot=True,cmap="RdBu", center=0)
# plt.title("Best Weight Values per Evolution")
plt.xlabel("Evolution")
plt.ylabel("Eval. Function")
# for label in (g.get_xticklabels() + g.get_yticklabels()):
# 	label.set_fontsize(14)
plt.show(g)
```

```python
sns.barplot(data=draw_df, x="function", y="value", hue="evolution")
```
