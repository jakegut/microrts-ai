```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("tourney_1619584012604.csv", skipinitialspace=True)
# df = df[df['AI2'] != 'CoacAI']
df = df[df['AI2'] != 'WorkerRush']
df = df[df['AI2'] != 'WorkerDefense']
df['fitness'] = -df['fitness']
print(df.head())
```

```python

sns.set_theme(font_scale=1.5, style="whitegrid")
plt.figure(figsize=(16,10))
g = sns.barplot(x="AI2", y="fitness", hue="map", data=df)
# plt.title("CMA-ES MCTS AI vs. multiple AI")
plt.ylabel("Fitness")
plt.xlabel("AI")
# plt.xticks(rotation=45)
# for label in (g.get_xticklabels() + g.get_yticklabels()):
# 	label.set_fontsize(12)
plt.show(g)
```
