```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file = open("fitness-final.txt", "r")
fitness = []
for _ in range(50):
    values = [float(i) for i in file.readline().strip().split(",")[1:]]
    fitness.append(values)
file.close()
```

```python
df = pd.DataFrame(fitness)
agg = pd.DataFrame()
agg['mean'] = df.mean(axis=1)
agg['max'] = df.max(axis=1)
agg['min'] = df.min(axis=1)
sns.boxplot(data=df.T)
```

```python
fig = sns.lineplot(data=agg)
plt.xlabel("Iteration")
plt.ylabel("Fitness")
plt.title("CMA-ES MCTS Fitness over 50 iterations")
plt.show(fig)
```

```python
draw_df = df.reset_index().melt(id_vars=['index'], var_name='col')
draw_df
```

```python
plt.figure(figsize=(18,9))
g = sns.lineplot(x="index", y="value", data=draw_df, ci=None)
plt.xlabel("Iteration", fontsize=16)
plt.ylabel("Fitness", fontsize=16)
plt.title("Fitness over time", fontsize=20)
for label in (g.get_xticklabels() + g.get_yticklabels()):
	label.set_fontsize(14)
plt.show(g)
```
