```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file = open("fitness-1618956350527.txt", "r")
file.readline()
fitness = []
for _ in range(50):
    values = [-1*float(i) for i in file.readline().strip().split(", ")[1:]]
    fitness.append(values)
file.close()
```

```python
df = pd.DataFrame(fitness)
agg = pd.DataFrame()
agg['mean'] = df.mean(axis=1)
agg['max'] = df.max(axis=1)
agg['min'] = df.min(axis=1)
sns.boxplot(data=df.T)
```

```python
fig = sns.lineplot(data=agg)
plt.xlabel("Iteration")
plt.ylabel("Fitness")
plt.title("CMA-ES MCTS Fitness over 50 iterations")
plt.show(fig)
```

```python
draw_df = df.reset_index().melt(id_vars=['index'], var_name='col')
draw_df.head()

plt.figure(figsize=(12,6))
sns.lineplot(x="index", y="value", data=draw_df)
```
