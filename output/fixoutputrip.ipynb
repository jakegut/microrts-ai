```python
import math

with open("output-1.txt", "r") as file:
    lines = file.readlines()
    
fitness = []
for line in lines:
    if line[0] == '[':
        temp = line.replace("[", "").strip()
        temp = temp.replace("]", "")
        temp = temp.split(", ")
        temp = math.prod([float(t) for t in temp])
        fitness.append(temp)
        
fitness.pop()
for it in range(50):
    print(str(it + 1) + "," + ",".join([str(i) for i in fitness[it:it+10]]))
```
